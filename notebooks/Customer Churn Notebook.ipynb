### **Defining Target and Droping Identifier**

```python
df=df.drop(columns=['customerID'])
y=df['Churn'].map({'No':0,'Yes':1})
x=df.drop(columns=['Churn'])
```

 ### **Split columns (numeric vs categorical)**

```python
num_cols=['tenure','MonthlyCharges','TotalCharges']
cat_cols=[c for c in x.columns if c not in num_cols]
```

### **Preprocessing Pipeline**

```python
# Imports
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, classification_report,precision_recall_curve, confusion_matrix
import joblib
import json
```

```python
preprocess = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
    ]
)
```

### **Split Data (Traning Data/ Testing Data)**

```python
X_train, X_test, Y_train, Y_test = train_test_split( x,y, test_size=0.2, random_state=42, stratify=y)
```

## **Model Testing and Evaluation**

### **Train baseline model**

```python
clf=LogisticRegression(max_iter=2000)

model=Pipeline(steps=[('preprocess',preprocess), ('clf',clf)])

model.fit(X_train,Y_train)

Y_proba = model.predict_proba(X_test)[:,1]
Y_pred = model.predict(X_test)

print("ROC-AUC:", roc_auc_score(Y_test, Y_proba))
print(classification_report(Y_test, Y_pred))
```

```python

joblib.dump(model, 'D:\\Github\\Customer-Churn-Prediction-System\\models\\churn_pipeline.joblib')
```

### **Tune the decision threshold**

```python
Y_scores=model.predict_proba(X_test)[:,1]

precision, recall, thresholds = precision_recall_curve(Y_test,Y_scores)

target_recall=0.70
idx=np.where(recall >=target_recall)[0][-1]
t= thresholds[idx-1]

Y_pred_t=(Y_scores >= t).astype(int)
print('Chosen Threshold',t)
print(classification_report(Y_test, Y_pred_t))
```

### **Retrain Logistic Regression with class weights**

```python
clf_bal=LogisticRegression(max_iter=2000, class_weight='balanced')
model_bal = Pipeline(steps=[('preprocess',preprocess),('clf',clf_bal)])

model_bal.fit(X_train,Y_train)

Y_proba = model_bal.predict_proba(X_test)[:,1]
print('AUC: ',roc_auc_score(Y_test,Y_proba))
print(classification_report(Y_test,model_bal.predict(X_test)))
```

```python
X_train, X_temp, Y_train, Y_temp = train_test_split(x,y, test_size=0.4, random_state=42, stratify=y)
X_val, X_test, Y_val, Y_test = train_test_split( X_temp, Y_temp, test_size=0.5, random_state=42, stratify=Y_temp)
```

```python
model.fit(X_train, Y_train)

val_scores=model.predict_proba(X_val)[:,1]
```

### **Proper Train/Val/Test**

```python


precision, recall, thresholds = precision_recall_curve(Y_val,val_scores)

target_recall=0.70
mask=recall[1:]>=target_recall
t= thresholds[mask].max()

val_pred=(val_scores > t).astype(int)
print('Chosen Threshold',t)
print("VAL report :\n",classification_report(Y_val,val_pred))

test_scores=model.predict_proba(X_test)[:,1]
test_pred=(test_scores >= t).astype(int)
print('TEST report :\n',classification_report(Y_test,test_pred))
```
